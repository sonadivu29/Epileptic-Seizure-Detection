```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('../data/Epileptic_Seizure_Recognition.csv') 
```

```python
# Display the first few rows to understand the structure
df.head()
```

```python
# Check basic information about the dataset
df.info()
```

```python
# Describe the dataset with statistics
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(f"Missing values:\n{missing_values}")
```

```python
# Drop or impute missing values (example: drop rows with missing values)
df.dropna(inplace=True) 
```

```python
# Check for duplicates and remove them
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
df = df.drop_duplicates()
```

```python
# Display data types of each column
print("Data types:")
print(df.dtypes)
```

```python
# Summary statistics for each feature, based on data types (numerical, categorical)
print(df.describe(include=['object']))
```

```python
# Visualizing distributions of numerical features
numerical_columns = df.select_dtypes(include=[np.number]).columns
df[numerical_columns].hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Visualize correlations between numerical features using a heatmap (error)
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot to check relationships between numerical features
sns.pairplot(df[numerical_columns])
plt.suptitle("Pairwise Relationships between Numerical Features", y=1.02)
plt.show()
```

```python
# Boxplot to detect outliers (e.g., for a specific feature)
plt.figure(figsize=(10, 6))
sns.boxplot(x='some_feature', data=df)  # Replace 'some_feature' with actual feature name
plt.title("Boxplot for Some Feature")
plt.show()
```

```python
# Value counts for categorical data (if any)
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"Value counts for {col}:")
    print(df[col].value_counts())
    print("\n")
```

```python
# Checking class imbalance (for classification tasks)
if 'target_column' in df.columns:  # Replace with your target column name
    sns.countplot(x='target_column', data=df)
    plt.title("Class Distribution")
    plt.show()
```
